```python
import cv2
import numpy as np

img = cv2.imread('./images/someshapes.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("Input image", gray)
cv2.waitKey(0)

# minimize computation using threshold function
ret, threshold_ = cv2.threshold(gray, 127, 255, 1)

# extract contours
_, contours, hierarchy = cv2.findContours(threshold_.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)

# Iterating through the contours 
for cnt in contours:
    # get Approximation 
    accuracy = 0.01 * cv2.arcLength(cnt, True)
    approx = cv2.approxPolyDP(cnt, accuracy, True)
    
    if len(approx) == 3:
        shape_name = "Triangle"
        # to draw contour
        cv2.drawContours(img, cnt, 0, (0,255,0), -1)
        
        # find contour center to place text at the center
        M = cv2.moments(cnt)
        cx = int (M['m10']/
                 M['m00'])
        cy = int(M['m01']/
                M['m00'])
        cv2.putText(img, shape_name, (cx-50, cy), cv2.FONT_HERSHEY_DUPLEX, 1, (0,0,0), 1)
        
    elif len(approx) == 4:
        x, y, w, h = cv2.boundingRect(cnt)
        M = cv2.moments(cnt)
        cx = int (M['m10']/M['m00'])
        cy = int (M['m01']/M['m00'])
        
        # checking whether rectangle or square
        
        if abs(h-w) <= 3:
            shape_name = "Square"
            cv2.drawContours(img, cnt, 0, (0,255,0), -1)
            cv2.putText(img, shape_name, (cx-50, cy), cv2.FONT_HERSHEY_DUPLEX, 1, (0,0,0), 1)
            
        else:
            shape_name = "Rectangle"
            cv2.drawContours(img, cnt, 0, (0,255,0), -1)
            cv2.putText(img, shape_name, (cx-50, cy), cv2.FONT_HERSHEY_DUPLEX, 1, (0,0,0), 1)
    
    elif len(approx) == 10:
        shape_name = "Star"
        M = cv2.moments(cnt)
        cx = int (M['m10']/M['m00'])
        cy = int (M['m01']/M['m00'])
        cv2.drawContours(img, cnt, 0, (0,255,0), -1)
        cv2.putText(img, shape_name, (cx-50, cy), cv2.FONT_HERSHEY_DUPLEX, 1, (0,0,0), 1)
            
    elif len(approx) == 10:
        shape_name = "Circle"
        M = cv2.moments(cnt)
        cx = int (M['m10']/M['m00'])
        cy = int (M['m01']/M['m00'])
        cv2.drawContours(img, cnt, 0, (0,255,0), -1)
        cv2.putText(img, shape_name, (cx-50, cy), cv2.FONT_HERSHEY_DUPLEX, 1, (0,0,0), 1)
        

# displaying identified parts
cv2.imshow("Identified parts", img)
cv2.waitKey(0)

cv2.destroyAllWindows()
            
            

            
            
            
```
